# Librerías

```python
import pandas as pd
import datetime
global df
from sklearn.preprocessing import LabelEncoder
pd.set_option('display.max_columns', None)
```

```python
#Funciones
#Carga de datos
def load_data(ruta,encoding,sep):
  global df
  df = pd.read_csv(ruta,encoding=encoding,sep=sep)

#Eliminar columna
def delete_column(dataframe,delete_colum):
  return dataframe.drop(columns=[delete_colum],inplace=True)

#Crear fecha con formato especifico
def create_date(dataframe,anio,cod_mes):
  dataframe['FECHA'] = pd.to_datetime(dataframe[anio].astype(str) + "-" +dataframe[cod_mes].astype(str) + "-01", format="%Y-%m-%d")
```

```python
#Usamos la función de carga de datos
load_data('/content/bd_museo_views.csv','ISO-8859-1',';')
#Mostramos primera fila
df.head(1)
```

```python
#Eliminamos columnas que no usaremos
delete_column(df,'TIPO_cat_cod')
delete_column(df,'FECHA_CORTE')
#Creamos columna de fecha
create_date(df,'ANIO','COD_MES')
```

```python
#Creamos columna de ADULTOS que hayan visitado el museo
df['ADULTOS'] = df['ADU_BOLESPPAGANTES'] + df['ADU_PAGANTES'] + df['ADU_BOLESPNOPAGANTES'] + df['ADU_NOPAGANTES']

#Creamos columna de ESTUDIANTES que hayan visitado el museo
df['ESTUDIANTE'] = df['EST_BOLESPPAGANTES'] + df['EST_PAGANTES'] + df['EST_BOLESPNOPAGANTES'] + df['EST_NOPAGANTES']

#Creamos columna de NIÑOS que hayan visitado el museo
df['NIÑOS'] = df['MIL_BOLESPPAGANTES'] + df['NIN_PAGANTES'] + df['NIN_BOLESPNOPAGANTES'] + df['NIN_NOPAGANTES']

#Creamos columna de ADULTOS_MAYORES que hayan visitado el museo
df['ADULTO_MAYOR'] = df['ADM_BOLESPPAGANTES'] + df['ADM_BOLESPNOPAGANTES']

#Creamos columna de MULITARES que hayan visitado el museo
df['MILITAR'] = df['MIL_BOLESPPAGANTES'] + df['MIL_BOLESPNOPAGANTES']
```

```python
#Crearemos una valor único para cada museo, categorica a numerico
le = LabelEncoder()
df['IDMUSEO'] = le.fit_transform(df['NOM_MUSEO'])

#Comprobamos que a cada museo se le asigno un número entero
indice_museo = df[['IDMUSEO','NOM_MUSEO']]
indice_museo.set_index('IDMUSEO', inplace=True)
indice_museo_unicos=indice_museo.drop_duplicates()
indice_museo_unicos.head()
```

```python
#Cambiamos el formato de un dataframe a un csv
df.to_csv('/content/sample_data/visita_museos.csv', index=False, header=True)
```
